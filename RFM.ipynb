## About The Dataset


This dataset represents a series of transactions recorded in an invoice log. Each row corresponds to a unique transaction and contains information such as the invoice number, stock code, description of the item, quantity purchased, date of the invoice, unit price, customer ID, and country of the customer.

The transactions appear to be related to retail sales of various items, including napkins, children's aprons, children's cutlery sets, and a baking set. The transactions all occurred on December 9, 2011, at 12:50 PM. The customer associated with these transactions has an ID of 12680.0 and is located in France.

### RFM analysis is a technique used in marketing and customer relationship management to segment and analyze customers based on their recent purchasing behavior. RFM stands for:

1. **Recency**: How recently a customer has made a purchase. It is often measured in terms of the time since the last purchase.

2. **Frequency**: How often a customer makes a purchase within a specific period, such as a month or a year.

3. **Monetary**: The total amount of money a customer has spent within a specific period.

## Recency
We already got the age of invoice 

```python
#recenccy

recency = maxdate.drop('InvoiceDate', axis=1)
recency
```

```python
freq=df.drop_duplicates(subset="InvoiceNo")
freq
```

```python
freq = freq.groupby(["CustomerID"]).count()[["InvoiceNo"]]
freq
```

## Monetary
Total amount that a customer expences

```python
#monetary
df["total"]=df["Quantity"]*df["UnitPrice"]
```

#### Now the total RFM comes out

```python
RFM=pd.concat([recency,freq,money], axis=1)
RFM
```

```python
RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

### We already got the RFM now it's time to cluster
#### Implemented K-Means Machine Learning Model To Cluster/Segment The Customers

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(RFM)
```

```python
scaled
```

```python
scaled.shape
```

```python
from sklearn.cluster import KMeans
```

### To find out the value of K

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k) 
    km.fit(scaled)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range, sse, 'r-*')
plt.xlabel ='k'
plt.ylabel = 'Sum of Squared Error'
```

### Object Creation and data fitting

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
kmeans.labels_+1
```

```python
RFM["Clusters"]=(kmeans.labels_+1)
```

#### Let's segment customers

```python
def func(row):
    if row["Clusters"]==3:
        return 'Dimond'
    elif row["Clusters"]==1:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['group']=RFM.apply(func, axis=1)
```

### RFM Segmentation Matrix

```python
RFM['R_Score'] = pd.qcut(RFM['Recency'], q=3, labels=['High', 'Medium', 'Low'])
RFM['F_Score'] = pd.qcut(RFM['Frequency'], q=3, labels=['Low', 'Medium', 'High'])
RFM['M_Score'] = pd.qcut(RFM['Monetary'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
pd.qcut(RFM['Recency'], q=3, labels=['High', 'Medium', 'Low'])
```

```python
# Combine RFM scores
RFM['RFM_Segment'] = RFM['R_Score'].astype(str) + RFM['F_Score'].astype(str) + RFM['M_Score'].astype(str)
```

```python
# Calculate RFM score
RFM['RFM_Score'] = RFM[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)
```

#### R_score x F_score

```python
# Create RFM segmentation matrix
rf_matrix = RFM.groupby(['R_Score', 'F_Score']).size().unstack()
sns.heatmap(rfm_matrix ,annot=True,fmt="d",cmap="Blues")
```

#### R_score x M_score

```python
rm_matrix = RFM.groupby(['R_Score', 'M_Score']).size().unstack()
sns.heatmap(rm_matrix ,annot=True,fmt="d",cmap="Blues")
```

#### F_score x M_score

```python
rm_matrix = RFM.groupby(['F_Score', 'M_Score']).size().unstack()
sns.heatmap(rm_matrix ,annot=True,fmt="d",cmap="Blues")
```

## RFM Heatmap
Each cell in the heatmap represents a combination of RFM scores, and the color intensity indicates the frequency or monetary value associated with that combination. This helps identify segments of customers with different RFM characteristics.
